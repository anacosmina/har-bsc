```
f = open("blah.txt", "r")
l = [[int(num) for num in line.split(',')] for line in f ]
```

A depth value is invalid if it is equal to 0 (i.e. the sensor has no data on this pixel).
Depth values equal to 0 are coloured in black.

```
import matplotlib.pyplot as plt

# See https://matplotlib.org/tutorials/colors/colormaps.html for more colormaps.
plt.matshow(l, cmap=plt.cm.gist_gray)

plt.show()
```
